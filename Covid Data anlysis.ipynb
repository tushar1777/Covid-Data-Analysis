# Relating the variable with scatterplots

### Graphical Representation of people recovered VS total number of cases

```python
sns.relplot(x="total_cases",y="recovered",data=data)
```

### Graphical Representation of people's death VS total number of cases

```python
sns.relplot(x="total_cases",y="deaths",data=data)
```

### Graphical Representation of total confirmed cases VS total number of cases

```python
sns.relplot(x="total_cases",y="total_confirmed_cases",data=data)
```

### Graphical Representation of people hospitalized with symptoms VS total number of cases

```python
sns.relplot(x="total_confirmed_cases",y="hospitalized_with_symptoms",hue='recovered',data=data)
```

### Graphical Representation of all the pairs of columns (Can be accessed on double clicking on the graph)

```python
sns.pairplot(data)
```

### Graphical Representation of people who had home quarantined VS total number of cases

```python
sns.relplot(x='total_cases',y='home_quarantine',kind='line',data=data)
```

### Graphical Representation of recovered people VS total number of cases

```python
sns.scatterplot(x='total_cases',y='recovered',data=data)
```
